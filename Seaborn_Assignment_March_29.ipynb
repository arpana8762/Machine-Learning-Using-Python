Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Solution:
Seaborn provides various plots to visualize data effectively. Here are five commonly used plots:

Line Plot (sns.lineplot())

Use: Used for visualizing trends over time or a continuous variable.
Example

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("fmri")
sns.lineplot(x="timepoint", y="signal", data=data)
plt.show()
```

Bar Plot (sns.barplot())

Use: Used to compare categorical data using bars.
Example

```python
data = sns.load_dataset("titanic")
sns.barplot(x="class", y="fare", data=data)
plt.show()
```

Box Plot (sns.boxplot())

Use: Displays the distribution of a numerical variable and identifies outliers.
Example:

```python
sns.boxplot(x="class", y="age", data=data)
plt.show()
```

Histogram (sns.histplot())

Use: Shows the frequency distribution of numerical data.
Example

```python
data = sns.load_dataset("diamonds")
sns.histplot(data["price"], bins=30)
plt.show()
```

Heatmap (sns.heatmap())

Use: Used for correlation matrices or visualizing tabular data using colors.
Example

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

# Compute the correlation matrix
corr_matrix = titanic.corr(numeric_only=True)  # Ensure only numeric columns are used

# Create the heatmap
plt.figure(figsize=(10,6))  # Adjust figure size for better readability
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Display the plot
plt.title("Correlation Matrix of Titanic Dataset")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Flights dataset
flights = sns.load_dataset("flights")

# Pivot the dataset for a heatmap
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

# Create a heatmap
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="coolwarm")

# Display the plot
plt.title("Number of Passengers (1949-1960)")
plt.show()
```

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Create a line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)

# Display the plot
plt.title("FMRI Signal over Time for Different Events and Regions")
plt.show()
```

Explanation:

hue="event" differentiates the lines based on event types.
style="region" gives different line styles based on region.
The fmri dataset tracks brain activity signals over time.

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Box plot for Age
sns.boxplot(x="pclass", y="age", data=titanic, ax=axes[0])
axes[0].set_title("Age Distribution Across Passenger Classes")

# Box plot for Fare
sns.boxplot(x="pclass", y="fare", data=titanic, ax=axes[1])
axes[1].set_title("Fare Distribution Across Passenger Classes")

# Display the plot
plt.show()
```

Explanation:

Two box plots are created to show the distribution of age and fare for each class.
Box plots help identify outliers and the spread of data

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
# Load the Diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Create a histogram
sns.histplot(diamonds, x="price", hue="cut", bins=30, kde=True)

# Display the plot
plt.title("Distribution of Diamond Prices by Cut")
plt.show()
```

Explanation:

The histogram shows how the price of diamonds varies.
hue="cut" colors the bars based on the cut type.
kde=True adds a density estimation curve.

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Flights dataset
flights = sns.load_dataset("flights")

# Pivot the dataset correctly
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

# Create a heatmap
plt.figure(figsize=(10, 6))  # Optional: Adjust figure size for better visibility
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="coolwarm")

# Display the plot
plt.title("Number of Passengers (1949-1960)")
plt.show()
```

Explanation:

The dataset is pivoted so that months become rows, years become columns, and passenger numbers are values.
The heatmap visually represents the number of passengers over the years.

```python
# Create a heatmap
plt.figure(figsize=(12, 6))  # Adjust figure size for better readability
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="coolwarm", linewidths=0.5)

# Display the plot
plt.title("Number of Passengers (1949-1960)")
plt.show()
```
